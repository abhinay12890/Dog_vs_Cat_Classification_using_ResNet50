# Dog vs Cat Classification using ResNet50

## Importing Libraries

```python
import pandas as pd
import os 
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
train_path='dog_cat_dataset/train' # path to data
test_path='dog_cat_dataset/test'
```

```python
lab=os.listdir(train_path)
lab
```

```python
all_train_images=[]
for x in lab:
    all_train_images.append(train_path+"/"+x)
all_train_images
```

```python
lab=os.listdir(test_path)
lab
```

```python
all_test_images=[]
for x in lab:
    all_test_images.append(test_path+"/"+x)

all_test_images
```

```python
train_df=pd.DataFrame({"image_path":all_train_images}) # converted all  train image paths into dataframe
train_df
```

```python
train_df['image_path'][0]
```

```python
os.path.basename(train_df['image_path'][0]).split(".")[0]
```

```python
def label_extract(x):
    return os.path.basename(x).split(".")[0]
```

```python
train_df['label']=train_df['image_path'].apply(label_extract) # adding labels for classification
train_df
```

```python
train_df['label'].value_counts()
```

Dog: 1, Cat:0

```python
train_df['label'].replace({"dog":1,"cat":0},inplace=True) # converting into integers
train_df
```

```python
paa=train_df[train_df["label"]==1]["image_path"].values[0] # reading sample dog image

i=Image.open(paa)
plt.imshow(i)
plt.title("Dog")
```

```python
paa=train_df[train_df["label"]==0]["image_path"].values[0] # reading sample cat image

i=Image.open(paa)
plt.imshow(i)
plt.title("Cat")
```

```python
test_df=pd.DataFrame({"image_path":all_test_images}) # adding all test image paths into dataframe
test_df 
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_df['label']=train_df['label'].astype(str)
```

## Reading Images from image path in dataframe

```python
# Performing Data augmentation
batch_size=8
train_datagen=ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    horizontal_flip=True,
    validation_split=0.3,
)

train_data=train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="label",
    target_size=(112,112),
    batch_size=batch_size,
    class_mode="binary",
    shuffle=True,
    subset="training")

val_data=train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col="image_path",
    y_col="label",
    target_size=(112,112),
    batch_size=batch_size,
    class_mode="binary",
    subset="validation",
    shuffle=False
)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
test_data=test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col="image_path",
    y_col=None,
    class_mode=None,
    target_size=(112,112),
    batch_size=batch_size,
)
```

```python
# Computing Class weights
from sklearn.utils.class_weight import compute_class_weight

classes=np.array(['1','0'])
y=train_df['label'].values

class_weight=compute_class_weight("balanced",classes=classes,y=y)
class_weight=dict(zip(classes,class_weight))

class_weight
```

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
```

### Loading Pre-Trained ResNet50 model

```python
# Loading Pre-Trained Model
base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(112, 112, 3))

# Freeze base layers
for layer in base_model.layers:
    layer.trainable = False
print(len(base_model.layers))
```

```python
# Training last few layers of the Model
for layer in base_model.layers[:-25]:
    layer.trainable=True
```

### Fine Tuning the output layer

```python
# Fine Tuning the model for this custom dataset
model=Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dropout(0.5))
model.add(Dense(1,activation="sigmoid"))
```

```python
print(model.summary())
```

```python
from tensorflow.keras.optimizers import Adam
# smaller LR for stable training
optimizer = Adam(learning_rate=1e-4)
```

```python
model.compile(optimizer=optimizer,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
# Implementing Regularization Techniques

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau


earlystopping=EarlyStopping(patience=10,monitor="val_accuracy",restore_best_weights=True)

lr_scheduler = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=5,
    min_lr=1e-6,
    verbose=1
)
```

```python
print(train_data.class_indices)
print(val_data.class_indices)
```

```python
class_weight = {0: 1.0, 1: 1.0}
```

```python
history=model.fit(train_data,epochs=50,class_weight=class_weight,validation_data=val_data,callbacks=[earlystopping,lr_scheduler])
```

```python
y_pred=model.predict(val_data)
y_pred
```

```python
y_pred_prob=y_pred.ravel()
y_pred_prob
```

## Evaluation Graphs

```python
import matplotlib.pyplot as plt

plt.plot(history.history["accuracy"],label="Training accuracy")
plt.plot(history.history["val_accuracy"],label="Validation accuracy")
plt.xlabel("Epochs")
plt.title("Accuracy")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["loss"],label="Training loss")
plt.plot(history.history["val_loss"],label="Validation loss")
plt.xlabel("Epochs")
plt.title("Loss")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["learning_rate"],label="Learning rate")
plt.xlabel("Epochs")
plt.ylabel("Learning Rate")
plt.title("Learning Rate Optimization")
plt.legend()
plt.show()
```

```python
y_true = val_data.classes

from sklearn.metrics import roc_curve, auc,confusion_matrix
fpr, tpr, _ = roc_curve(y_true, y_pred_prob)
roc_auc = auc(fpr, tpr)
print("Validation ROC AUC:", roc_auc)
```

```python
plt.plot(fpr,tpr,color="blue")
plt.plot([0,1],[0,1],linestyle="--",color='red')
plt.xlim([0.0,0.1])
plt.ylim([0.0,1.05])
plt.xlabel("False Positive rate")
plt.ylabel("True Postive Rate")
plt.title('ROC Curve')
plt.legend()
plt.show()
```

```python
y_labels=np.where(y_pred_prob<0.5,0,1)
y_labels
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true,y_labels))
```

```python
import seaborn as sns
cm=confusion_matrix(y_true,y_labels)

sns.heatmap(cm,annot=True,cmap="Blues",fmt=".2f")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```
